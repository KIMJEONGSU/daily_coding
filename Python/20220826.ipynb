```python
####첫번째 코드
a=list(map(int,input().split()))
a.sort()
result=0

if a[0]==a[1]==a[2]:    #주사위 같은 눈 3개일때
    result=10000+(a[0]*1000)
    
elif a[0]!=a[1] and a[0]!=a[2] and a[1]!=a[2]: #주사위 같은 눈 0개 일때
    result=a[2]*100
    
elif a[0] in (a[1],a[2]):
    result=1000+(a[0]*100)
    
elif a[1] in (a[2]):
    result=1000+(a[1]*100)
    

print(result)

####런타임에러
```

```python
#####두번째 코드
a,b,c=map(int,input().split())

if a==b==c:    #주사위 같은 눈 3개일때
    result=10000+(a*1000)
    
elif a==c: #주사위 같은 눈 0개 일때
    result=1000+(a*100)
    
elif a==b:
    result=1000+(a*100)
    
elif b==c:
    result=1000+(b*100)
else:
    result=max(a,b,c)*100
    

print(result)
```
