```python
n= int(input())
num=n
a=0

while True:
    x=num//10   #10의자리수
    y=num%10  #1의자리수
    c=(x+y) % 10
    num=(y*10)+c
    
    a+=1
    if num==n:
        break
    
print(a)

#관련 함수를 찾아서 코드작성하려고 했지만 사칙연산 사용
```
