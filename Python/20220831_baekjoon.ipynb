```python
n=int(input())

for i in range(n):
    a=list(map(int,input().split()))
    sum_j=sum(a[1:])/a[0]
    count=0 # 평균보다 높은 학생 카운트
    for j in a[1:]:
        if j>sum_j:
            count+=1
    result=count/a[0]*100      #평균값보다 높은 점수가 있으면 count +1
    print(f'{result:.3f}%')    #소수점 3자리까지 출력하고 %붙여주기
```

```python
def solve(a):
    ans = 0
    for i in a:
        ans+=i
    return ans
```
