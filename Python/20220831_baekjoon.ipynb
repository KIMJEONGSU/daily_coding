```python
n=int(input())

for i in range(n):
    a=list(map(int,input().split()))
    sum_j=sum(a[1:])/a[0]
    count=0 #    
    for j in a[1:]:
        if j>sum_j:
            count+=1
    result=count/a[0]*100      #    count +1
    print(f'{result:.3f}%')    # 3  %
```

```python
def solve(a):
    ans = 0
    for i in a:
        ans+=i
    return ans
```
