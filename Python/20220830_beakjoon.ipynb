```python
#나의 풀이
a=[]     #빈리스트
for i in range(9):    #9개의 숫자를 
    n=int(input())     #줄마다 입력을 받아
    a.append(n)        #빈 리스트에 넣어준다.    
    
    
m=a[0]                 #a[0]를 기준으로
for i in range(1,len(a)):   #인덱스 1부터 8까지 비교
    if m<a[i]:               #기준 숫자보다 크면 큰숫자를 기준으로 (반복)
        m=a[i]

print(m,a.index(m)+1)     #인덱스는 0부터 시작하니까 +1해서 몇번째인지 출력한다.
```

```python
#남의 풀이
num_list = []
for i in range(9):
    num_list.append(int(input()))
    
print(max(num_list))        #max함수 사용
print(num_list.index(max(num_list))+1)
```

```python
a = [] 
for i in range(10):
    n=int(input())
    m=n%42
    a.append(m)

#빈리스트를 만들고 42로 나눈 나머지를 넣도록 for반복분을 실행
    
print(len(set(a))) 
#set을 사용해서 중복된 값은 없애서 len으로 리스트의 길이를 출력
```

```python
n=int(input())  #과목개수
m=list(map(int,input().split()))   #점수 입력받기

m.sort()   #과목 정렬

for i in range(len(m)):
    m[i]=m[i]/m[-1]*100  #낮은점수를 새로 계산한 값


print(sum(m)/len(m))
```

```python
n=int(input())  #몇 줄인지 입력


for i in range(n):
    m=list(input())   #n번동안 리스트 입력받기
    score=0           
    result=0
    for j in m:
        if j=='O':
            score+=1      #'O'가 연속되면 1씩 증가하기
            result+=score    #1씩증가된 값 더하기
        else:
            score=0          #'X'가 나오면 score를 0으로 만들기
            
    print(result)
    
    
#문제에 대한 코딩해답
```
